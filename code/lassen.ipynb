# Step 1: Import lassen_analysis script

```python
import sys
sys.path.append('./code')
import lassen_analysis as la
```

# Step 2: Load Data

```python
df_eruptions_C00_D100_P00, df_summary_C00_D100_P00 = la.load_run_data("C00_D100_P00_1Bar")

df_eruptions_C00_D00_P100, df_summary_C00_D00_P100 = la.load_run_data("C00_D00_P100_1Bar")

df_eruptions_C100_D00_P00, df_summary_C100_D00_P00 = la.load_run_data("C100_D00_P00_1Bar")

df_eruptions_C33_D33_P33, df_summary_C33_D33_P33 = la.load_run_data("C33_D33_P33_1Bar")

df_eruptions_C00_D100_P00_FC, df_summary_C00_D100_P00_FC = la.load_run_data("C00_D100_P00_1Bar_FC")

df_eruptions_C00_D00_P100_FC, df_summary_C00_D00_P100_FC = la.load_run_data("C00_D00_P100_1Bar_FC")

df_eruptions_C100_D00_P00_FC, df_summary_C100_D00_P00_FC = la.load_run_data("C100_D00_P00_1Bar_FC")

df_eruptions_C33_D33_P33_FC, df_summary_C33_D33_P33_FC = la.load_run_data("C33_D33_P33_1Bar_FC")
```

# Step 3: Clean Data

```python
df_eruptions_C00_D100_P00 = la.clean_settp_duplicates(df_eruptions_C00_D100_P00)

df_eruptions_C00_D00_P100 = la.clean_settp_duplicates(df_eruptions_C00_D00_P100)

df_eruptions_C100_D00_P00 = la.clean_settp_duplicates(df_eruptions_C100_D00_P00)

df_eruptions_C33_D33_P33 = la.clean_settp_duplicates(df_eruptions_C33_D33_P33)

df_eruptions_C00_D100_P00_FC = la.clean_settp_duplicates(df_eruptions_C00_D100_P00)

df_eruptions_C00_D00_P100_FC = la.clean_settp_duplicates(df_eruptions_C00_D00_P100)

df_eruptions_C100_D00_P00_FC = la.clean_settp_duplicates(df_eruptions_C100_D00_P00)

df_eruptions_C33_D33_P33_FC = la.clean_settp_duplicates(df_eruptions_C33_D33_P33)
```

# Step 4a: Plot Hydrous Oxides

```python
df_list = [df_eruptions_C00_D100_P00, df_eruptions_C00_D00_P100, df_eruptions_C100_D00_P00, df_eruptions_C33_D33_P33,
          df_eruptions_C00_D100_P00_FC, df_eruptions_C00_D00_P100_FC, df_eruptions_C100_D00_P00_FC, df_eruptions_C33_D33_P33_FC]
run_names = ['C00_D100_P00', 'C00_D00_P100', 'C100_D00_P00', 'C33_D33_P33',
            'C00_D100_P00_FC', 'C00_D00_P100_FC', 'C100_D00_P00_FC', 'C33_D33_P33_FC']

fig = la.plot_all_oxides_vs_pressure(df_list, run_names)

la.save_fig(fig, 'Hydrous-Oxides.png')
```
