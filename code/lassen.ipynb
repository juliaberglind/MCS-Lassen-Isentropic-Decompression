# Step 1: Import lassen_analysis script

```python
import sys
sys.path.append('./code')
import lassen_analysis as la
```

# Step 2: Load Data

```python
df_eruptions_C00_D100_P00, df_summary_C00_D100_P00 = la.load_run_data("C00_D100_P00")

df_eruptions_C00_D00_P100, df_summary_C00_D00_P100 = la.load_run_data("C00_D00_P100")

df_eruptions_C100_D00_P00, df_summary_C100_D00_P00 = la.load_run_data("C100_D00_P00")

df_eruptions_C33_D33_P33, df_summary_C33_D33_P33 = la.load_run_data("C33_D33_P33")
```

# Step 3: Clean Data

```python
df_eruptions_C00_D100_P00 = la.clean_settp_duplicates(df_eruptions_C00_D100_P00)

df_eruptions_C00_D00_P100 = la.clean_settp_duplicates(df_eruptions_C00_D00_P100)

df_eruptions_C100_D00_P00 = la.clean_settp_duplicates(df_eruptions_C100_D00_P00)

df_eruptions_C33_D33_P33 = la.clean_settp_duplicates(df_eruptions_C33_D33_P33)
```

# Step 4: Plot Oxides

```python
df_list = [df_eruptions_C00_D100_P00, df_eruptions_C00_D00_P100, df_eruptions_C100_D00_P00, df_eruptions_C33_D33_P33]
run_names = ['C00_D100_P00', 'C00_D00_P100', 'C100_D00_P00', 'C33_D33_P33']

fig = la.plot_all_oxides_vs_pressure(df_list, run_names)

la.save_fig(fig, 'oxides_end_members.png')
```

# Step 5: Plot Temperature vs. Pressure

```python
fig = la.plot_temperature_vs_pressure(df_list, run_names)

la.save_fig(fig, 't-p_end_members.png')
```

# Step 6: Plot fluid mass versus pressure and temperature

```python
fig = la.plot_fluid_mass(df_list, run_names)
la.save_fig(fig, 'fluid-mass_end_members.png')
```

```python
fig = la.plot_pressure_crystallinity(df_list, run_names)
la.save_fig(fig, 'crystallinity-pressure_end_members.png')
```
